```python
from datasets import load_dataset

# 1. Cargar el dataset de Common Voice en español
# Get presigned URL and download file
from datacollective import download_dataset
dataset_path = download_dataset("cmqim2spa00synr071fcp7av0")

from datacollective import load_dataset
# Load as a Pandas DataFrame, if supported!
dataset = load_dataset("cmqim2spa00synr071fcp7av0")
```

```python
# 2. Filtrar específicamente por acento de España
dataset_es_es = dataset.filter(lambda x: x["accent"] == "España" or x["accent"] == "Spain")
dataset_es_es
```
