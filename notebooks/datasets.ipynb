```python
import requests
import urllib3

urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

_orig = requests.sessions.Session.request

def _request_no_verify(self, method, url, **kwargs):
    kwargs.setdefault("verify", False)
    return _orig(self, method, url, **kwargs)

requests.sessions.Session.request = _request_no_verify
```

```python
# 2. Filtrar específicamente por acento de España
dataset_es_es = dataset.filter(lambda x: x["accent"] == "España" or x["accent"] == "Spain")
dataset_es_es
```
